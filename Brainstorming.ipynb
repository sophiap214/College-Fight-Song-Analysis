```python
tropes = ['fight','victory','win_won','rah','nonsense','colors','men','opponents','spelling']
data[tropes] = data[tropes].replace({'Yes': 1, 'No': 0})
data[tropes].mean().sort_values().plot(kind='barh')
```

```python
df = data[data['year'] != 'Unknown'].copy()

# Convert year to integer
df['year'] = df['year'].astype(int)

# Create decade column
df['decade'] = (df['year'] // 10) * 10

# Aggregate: proportion of songs mentioning men by decade
men_by_decade = df.groupby('decade')['men'].mean()

# Plot
plt.figure()
plt.plot(men_by_decade.index, men_by_decade.values, marker='o')
plt.xlabel('Decade')
plt.ylabel('Proportion of Songs Mentioning Men')
plt.title('Use of Gendered Language in Fight Songs by Decade')
plt.ylim(0, 1)
plt.show()
data = pd.read_csv('fight-songs.csv')

# Remove unknown years
df = data[data['year'] != 'Unknown'].copy()

# Convert year to integer
df['year'] = df['year'].astype(int)

# Create decade column
df['decade'] = (df['year'] // 10) * 10

# Count number of songs per decade
songs_by_decade = df.groupby('decade').size()

# Plot
plt.figure()
plt.bar(songs_by_decade.index, songs_by_decade.values)
plt.xlabel('Decade')
plt.ylabel('Number of Fight Songs')
plt.title('Frequency of Fight Songs by Decade')
plt.show()
```

```python
tropes = [
    'fight','victory','win_won','rah',
    'nonsense','colors','men','opponents','spelling'
]

# Convert to numeric safely
for col in tropes:
    data[col] = (data[col] == 'Yes').astype(int)


conference_tropes = data.groupby('conference')[tropes].mean()
plt.figure(figsize=(10, 6))
plt.imshow(conference_tropes, aspect='auto')
plt.colorbar(label='Proportion of Songs')
plt.xticks(range(len(tropes)), tropes, rotation=45, ha='right')
plt.yticks(range(len(conference_tropes.index)), conference_tropes.index)
plt.title('Frequency of Fight Song Tropes by Conference')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

complete = pd.read_csv("fight-songs.csv")



# convert year to a clean integer-like numeric
complete["year"] = (
    complete["year"]
    .astype(str)
    .str.extract(r"(\d{4})")[0]          # pulls a 4-digit year if it exists
    .astype("float")                    # allows NaN
)

complete = complete.dropna(subset=["year"])
complete["year"] = complete["year"].astype(int)

complete["decade"] = (complete["year"] // 10) * 10
complete["decade"] = (complete["year"] // 10) * 10

# Transforms yes/no into bool
complete["men_bool"] = (
    complete["men"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": True, "no": False})
)

complete["victory_win_won_bool"] = (
    complete["victory_win_won"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": True, "no": False})
)

complete["fight_bool"] = (
    complete["fight"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": True, "no": False})
)

#Calculates the proportion per decade
prop_men_by_decade = complete.groupby("decade")["men_bool"].mean()
prop_victory_win_won_by_decade = complete.groupby("decade")["victory_win_won_bool"].mean()
prop_fight_by_decade = complete.groupby("decade")["fight_bool"].mean()


#Plot 1
plt.figure()

prop_men_by_decade.plot(marker="o", label="Men")
prop_victory_win_won_by_decade.plot(marker="o", label="Victory/Win/Won")
prop_fight_by_decade.plot(marker="o", label="Fight")

plt.xlabel("Decade")
plt.ylabel("Proportion of songs")
plt.title("Mentions in Fight Songs by Decade")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os

font_path = os.path.join(os.getcwd(), "Clarendon Bold.otf")
fm.fontManager.addfont(font_path)

clarendon = fm.FontProperties(fname=font_path)
print(clarendon.get_name())


#Football field plot
decades = prop_men_by_decade.index.to_list()
x = np.arange(len(decades))

fig, ax = plt.subplots(figsize=(12, 6))

ax.set_facecolor("#35654d")
fig.patch.set_facecolor("#35654d")

for i in x:
    ax.axvline(i, color="white", alpha=0.3, linewidth=1)

for i in x[::5]:
    ax.axvline(i, color="white", alpha=0.7, linewidth=2)

ax.plot(x, prop_men_by_decade.values, marker="o", label="Men")
ax.plot(x, prop_victory_win_won_by_decade.values, marker="o", label="Victory/Win/Won")
ax.plot(x, prop_fight_by_decade.values, marker="o", label="Fight")

ax.set_xticks(x)
ax.set_xticklabels(
    [f"'{str(d)[-2:]}s" for d in decades],
    fontproperties=clarendon,
    fontsize=20, color = 'white'
)
ax.set_ylim(0, 1)

ax.set_xlabel("Decade (yard lines)")
ax.set_ylabel("Proportion of songs")
ax.legend()
plt.show()
```
